## - 2. Load the source code of the page from the Internet into this object

```python
from bs4 import BeautifulSoup
import urllib.request

fp = urllib.request.urlopen("https://tvtropes.org/pmwiki/pmwiki.php/Series/SusuASarkany")
soup = BeautifulSoup(fp)
print(soup)
print("End!!!")
```
