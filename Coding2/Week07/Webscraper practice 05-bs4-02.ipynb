## - 1. scrap the data of specific url

```python
import requests
if __name__ == "__main__":
    url = "https://tvtropes.org/pmwiki/pmwiki.php/Series/SusuASarkany"
    # point out the url
    response = requests.get(url=url)
    page_text = response.text
    #get the response object
    print(page_text)
    #data persistence store
    with open("./ greendragon-susu.html", "w",encoding="utf-8") as fp:
        fp.write(page_text)
    print("End of scraped data！！")
```

## - 2. parsing the specific data

```python
from bs4 import BeautifulSoup
jx01 = open("./ greendragon-susu.html")
soup = BeautifulSoup(jx01)
#print(soup)
#print(soup.a)
#print("!!end 01")
#print(soup.find("a"))
#print("!!end 02")
#print(soup.find("a",href="/"))
#print("!!end 03")
#print(soup.find("div",class_="titles"))
#print("!!end 04")
#print(soup.find_all("a"))
#print("!!end 05")
print(soup.find("div",class_="titles").text)
print("!!end 06")
print(soup.find("div",class_="sidebar-item sb-ad-unit").text)
print(soup.select(".sidebar-item > ul a")[0].get_text)
print("!!end 07")
print(soup.select(".sidebar-item > ul a")[0]["href"])
```
